```python
# growth_rate 2017


import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df_final= pd.read_csv('/Users/tiantianwu/Desktop/final_data.csv')
df_bubble = df_final[['country', 'year', 'growth_rate', 'population','birth_rate', 'income']][df_final['year']== 2017]
N = 142
x = df_bubble['birth_rate']
y = df_bubble['income']
z = 75*df_bubble['growth_rate']
colors = np.random.rand(N)
plt.scatter(x, y,z,c=colors, alpha=0.6,label='growth_rate')
plt.xlabel('Birth_rate(%)')
plt.ylabel('Income($)')
plt.title('Birth_rate-2017 & Income-2017 & growth_rate-2017')
plt.show()
```

```python
# growth_rate 2018


import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df_final= pd.read_csv('/Users/tiantianwu/Desktop/final_data.csv')
df_bubble = df_final[['country', 'year', 'growth_rate', 'population','birth_rate', 'income']][df_final['year']== 2017]
df_2018= pd.read_csv('/Users/tiantianwu/Desktop/visual_pred1.csv')
df_2018_growth_rate = df_2018[['country','Pred_2018_C0']]
ts_20_2018 = pd.merge(df_bubble, df_2018_growth_rate, how='left', on=['country'])
N = 142
x = ts_20_2018['birth_rate']
y = ts_20_2018['income']
z = 75*ts_20_2018['Pred_2018_C0']
colors = np.random.rand(N)
plt.scatter(x, y,z,c=colors, alpha=0.6,label='growth_rate_2018')
plt.xlabel('Birth_rate(%)')
plt.ylabel('Income($)')
plt.title('Birth_rate-2017 & Income-2017 & growth_rate-2018')
plt.show()


```
